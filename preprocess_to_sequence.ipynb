```python
import pandas as pd
train=pd.read_csv('./Datasets/amazon/train_amazon.csv',sep='\t')
train
```

```python
train.describe()
```

```python
train['user_id'].max()
```

```python
train_sort=train.sort_values('time')
```

```python
train_sort
```

```python
train_sort[train_sort.user_id==1]
```

```python
train_list=list()
for i in range(train['user_id'].max()):
    one_list=list(train_sort[train_sort.user_id==(i+1)].item_id.values-1)
    if(len(one_list)<=1):
        print(i,one_list)
    train_list.append(one_list)
```

```python
import pickle
fs=open("./Datasets/amazon/sequence", 'wb')
pickle.dump(train_list,fs)
fs.close()
```
